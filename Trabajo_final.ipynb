#### LIMPIEZA DE LA BASE DE DATOS

````python
# Cargando la información correspondiente
ind = pd.read_csv("./Indicadores_de_desarrollo.csv", encoding="latin1")
ind
````

En esta oportunidad no necesitaremos todos los países ni todos los años de los Indicadores de Desarrollo Global 
- Quedémonos con países pertenecientes a la Alianza del Pacífico  ```Chile Colombia Mexico Peru``` 
- Quedémonos con un periodo de tiempo de 7 años: 2010 - 2017

````python
# Hacemos una copia del dataframe 
idg = ind.copy()
idg = idg.set_index('Country Name')

# Nos quedamos con las variables solicitadas 
columnas = ["Country Code", "Series Name",'2010 [YR2010]', '2011 [YR2011]', '2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]',
            '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]']
filas = ["Chile", "Colombia", "Mexico", "Peru"]
new_idg = idg.loc[filas, columnas]
new_idg

# Cambiamos los títulos del dataframe
new_idg = new_idg.rename(columns={'2010 [YR2010]': '2010', '2011 [YR2011]': '2011', '2012 [YR2012]': '2012', 
                                  '2013 [YR2013]': '2013', '2014 [YR2014]': '2014', '2015 [YR2015]': '2015',
                                  '2016 [YR2016]': '2016', '2017 [YR2017]': '2017'
                                 })

# Guardamos nuestro nuevo data frame
new_idg.to_csv('New_IDG.csv')
````

##### Realizando un gráfico del crecimiento del PIB de Chile en el periodo 2010 - 2017

````python
# Cambiamos de formato de wide a long:
tbl_melt = base.melt( id_vars = ['Country Name', 'Country Code', 'Series Name'],
                   value_vars = [col for col in base if col.startswith('20')])
# Modificando nombres
new_tbl_melt = tbl_melt.rename(columns={'variable': 'Year', 'value': 'Value'})
````

````python
# Filtrar por 'Series Name' y 'Country Name'
con1 = new_tbl_melt['Country Name'] == 'Chile'
con2 = new_tbl_melt['Series Name'] == 'GDP growth (annual %)'

# Aplicar los filtros
resultado = new_tbl_melt[con1 & con2]
resultado
````

````python
# Graficando
chart = alt.Chart(resultado).mark_line().encode(
    x=alt.X('Year:O', title = "Año"),
    y=alt.Y('Value:Q', title = "Crecimiento del PIB"),
    tooltip=['Year:O', 'Value:Q']
).properties(
    width=340,  # Ajustar el ancho del gráfico
    height=240,  # Ajustar la altura del gráfico
    title="Crecimiento del PBI de Chile en el periodo 2010 - 2017"
)

# Mostrar el gráfico
chart.display()
````

##### Realizando un gráfico del crecimiento del PIB de los países miembros de la Alianza del Pacífico en el periodo 2010- 2017
